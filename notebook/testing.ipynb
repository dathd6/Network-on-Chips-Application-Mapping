### Hyperparameter Multi-objective Optimization for Support Vector Regression (SVR) Machine Learning Algorithm
3 objectives: Root Mean Squared Error measure
- MVP Share Train/Test RMSE 
- MVP Ranking RMSE

#### NSGA-II: 2 objectives Train RMSE and Test RMSE

```python
import numpy as np
nsga_ii = NSGA_II(
    nba_stats,
    pop_size=100,
    n_tournaments=35,
    n_evaluations=2000,
    mvp_share_rmse=True,
    execution_time=False,
    mvp_ranking_rmse=True,
)
pop = np.array(copy.deepcopy(nsga_ii.population))
```

```python
nsga_ii.optimize()
```

#### Function Visualize hypervolume metrics to compare optimisers

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

def visualise_perf(opt1, opt2):
    perf_metrics_1 = np.array(opt1.perf_metrics)
    perf_metrics_2 = np.array(opt2.perf_metrics)
    sns.lineplot(
        x=perf_metrics_1[:, 0],
        y=perf_metrics_1[:, 1],
        label='NSGA-II',
        color='#2377B4'
    )
    sns.lineplot(
        x=perf_metrics_2[:, 0],
        y=perf_metrics_2[:, 1],
        label='MOEA/D',
        color='#FA8010'
    )
    #plt.fill_between(
    #    x=x_range,
    #    y1=perf_metrics_1[:, 0],
    #    y2=perf_metrics_1[:, 2],
    #    color='#2377B4',
    #    alpha=0.2,
    #)
    #plt.fill_between(
    #    x=x_range,
    #    y1=perf_metrics_2[:, 0],
    #    y2=perf_metrics_2[:, 2],
    #    color='#FA8010',
    #    alpha=0.2,
    #)
    plt.xlabel('No. evaluations')
    plt.ylabel('Hypervolume')
    plt.title('HV convergence')
    plt.show()
    print('NSGA-II: ', perf_metrics_1[:, 1][-1])
    print('MOEA/D: ', perf_metrics_2[:, 1][-2])
```

```python
perf_1.append(nsga_ii.perf_metrics)
perf_2.append(moead.perf_metrics)
perf_3.append(nsga_ii_3D.perf_metrics)
perf_4.append(moead_3D.perf_metrics)
```
